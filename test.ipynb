```python
from Fluid_Mechanics.FluidSystems import System
from sympy import Symbol
```

```python
system = System()
```

```python
x = Symbol("x")
y = Symbol("y")
```

```python
system.u = -0.004 * (x /(x * x + y * y))
system.v = - 0.004 * (y /(x * x + y * y))
```

```python
summary = system.get_summary([[-2, 2],[-2, 2]], num=50)
summary
```
